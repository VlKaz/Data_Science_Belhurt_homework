```python
from mod import (linear, mser2, visual, clean_data, gbr, catboost,crossvalline,crossvalgbr,crossvalcatb)
#from colorama import Fore, Style, init

def main():
    filedata = 'data/data.csv'
    
    data = clean_data(filedata) 
    print(data.head())
    # Запускаем Линейную регрессию
    y_test,y_pred = linear(data) 
    # Считаем mse, r2, mae, mape для линейной регрессии
    mse_l, r2_l, mae_l, mape_l = mser2(y_test,y_pred)
    print('MSE_Linear:', mse_l)
    print('R²_Linear:', r2_l)
    print('MAE_Linear:', mae_l)
    print('MAPE_Linear:', mape_l)
    crossvalline(data)
    # Выводим график модели Линейной Регрессии
    visual(y_test,y_pred)
    print('--------------------------------------------------------------')
     # Запускаем GBR регрессию
    y_test1,y_pred1 = gbr(data)
    # Считаем mse, r2, mae, mape для GBR регрессии
    mse_l1, r2_l1, mae_l1, mape_l1 = mser2(y_test1,y_pred1)
    print('MSE_GBR:', mse_l1)
    print('R²_GBR:', r2_l1)
    print('MAE_GBR:', mae_l1)
    print('MAPE_GBR:', mape_l1)
    crossvalgbr(data)
    # Выводим график модели GBR Регрессии
    print('График GBR')
    visual(y_test1,y_pred1)
    print('--------------------------------------------------------------')
     # Запускаем CATB регрессию
    y_test2,y_pred2 = catboost(data)
     # Считаем mse, r2, mae, mape регрессии
    mse_l2, r2_l2, mae_l2, mape_l2 = mser2(y_test2,y_pred2)
    print('MSE_CATB:', mse_l2)
    print('R²_CATB:', r2_l2)
    print('MAE_CATB:', mae_l2)
    print('MAPE_CATB:', mape_l2)
    crossvalcatb(data)
     # Выводим график модели CATB Регрессии
    print('График CATBOOST')
    visual(y_test2,y_pred2)
       
if __name__ == "__main__":
    main()
```
