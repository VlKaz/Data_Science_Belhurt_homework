```python
from mod import (linear, mser2, visual, clean_data, gbr, catboost)
#from colorama import Fore, Style, init

def main():
    filedata = 'data/data.csv'
    
    data = clean_data(filedata) 
    print(data)
    # Запускаем Линейную регрессию
    y_test,y_pred = linear(data) 
    # Считаем MSE, R2 для линейной регрессии
    mse_l,r2_l = mser2(y_test,y_pred)
    print('MSE_Linear:', mse_l)
    print('R²_Linear:', r2_l)
    # Выводим график модели Линейной Регрессии
    visual(y_test,y_pred)
     # Запускаем GBR регрессию
    y_test1,y_pred1 = gbr(data)
    # Считаем MSE, R2 для GBR регрессии
    mse_l1,r2_l1 = mser2(y_test1,y_pred1)
    print('MSE_Linear:', mse_l1)
    print('R²_Linear:', r2_l1)
    # Выводим график модели GBR Регрессии
    visual(y_test1,y_pred1)

     # Запускаем CATB регрессию
    y_test2,y_pred2 = catboost(data)
     # Считаем MSE, R2 для CATB регрессии
    mse_l2,r2_l2 = mser2(y_test2,y_pred2)
    print('MSE_Linear:', mse_l2)
    print('R²_Linear:', r2_l2)
     # Выводим график модели CATB Регрессии
    visual(y_test2,y_pred2)
    #if mse_l < mse_a and r2_l > r2_a: 
    #    best_model = "Линейная регрессия" 
    #else: 
    #    best_model = "ARIMA"
    #print(Fore.GREEN + Style.BRIGHT + f"Лучшая модель: {best_model}")
    
if __name__ == "__main__":
    main()
```
