12.	Написать программу, которая принимает от пользователя два числа и выводит их произведение.

```python
a = input("Введите число a: ")
b = input("Введите число b: ")
mult = float(a)*float(b)
print("Результат умножения",mult)
print("Сложность этой операции О(1)")
```

Сложность это операции составляет О(1)
